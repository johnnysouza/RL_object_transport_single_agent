### Análise dos resultados do agente de Q Learning
Cenários para diversos valores diversos para o a taxa de aprendizagem (alfa), fator de desconto (gama) e política de exploração (épsilon)

```python
import matplotlib.pyplot as plt
import json
```

```python
with open('result.txt', 'r') as file:
    results = file.readlines()
    results = [json.loads(result) for result in results]
```

```python
for result in results:
    print(f"Valor ótimo de {min(result['steps_by_episode'])} passos")
```

```python
for result in results:
    fig, (plt1, plt2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    labels = range(len(result['steps_by_episode']))
    label_counts = result['steps_by_episode']

    plt1.set_title("Evolução de passos por período")
    plt1.set_xlabel("Periodos")
    plt1.set_ylabel("Passos por período")
    plt1.plot(labels, label_counts)

    labels = range(len(result['time_by_episode']))
    label_counts = result['time_by_episode']

    plt2.set_title("Evolução do tempo de execução por período")
    plt2.set_xlabel("Periodos")
    plt2.set_ylabel("Microsegundos por período")
    plt2.plot(labels, label_counts)

    fig.suptitle(f"Análise com alfa= {result['alpha']}, gama= {result['gamma']} e épsilon= {result['epsilon']}")
    plt.show()
```

```python
for result in results:
    episodies_with_optime_result = len([steps_by_episode for steps_by_episode in result['steps_by_episode'] if steps_by_episode == 8])
    print(f"Análise com alfa= {result['alpha']}, gama= {result['gamma']} e épsilon= {result['epsilon']}, quantidade de episódios que atingiram o mínimo de passos: {episodies_with_optime_result}")
```

Pode-se perceber muita oscilação com episódios que atingem o limite de 500 passos sem chegar na solução, acredita-se que tais variáções sejam explicadas pela ação de permanecer onde está, o que em alguns casos pode atrapalhar a evolução do agente.

Melhor cenário encontrado foi com alfa de 0.3, gama de 0.5 e épsilon de 0.1. Nele houve um rápido padrão de convergência para o valor ótimo de 8 passos em um episódio para o estado final, pouca variação com picos de 500 passos sem chega no estado final e a maior quantidade de epiódios que atingiram o valor ótimo de 8 passos.
